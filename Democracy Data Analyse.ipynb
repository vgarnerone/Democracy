```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt


# data frame with geographic data
df_GEO = pd.read_csv('DATA/private.csv')
# columns explicitation : 
    # "Country" : name of the country
    # "Democracy": 1 is a democracy, 0 a dictatorship
    # "Expectancy": age expectancy in year
    # "PopulationD" : density of population in hab/km2
    # "Gini" : coefficient de Gini, evaluates wealth inequality
    # "AgeMed" : median age of the population in year
    # "Skyscraper" : high the highest building in meters 
    # "Children": number of children per woman
    # "PressF" : press freedom score

# data frame with political data
df_POLI = pd.read_csv('DATA/Democracy-Index-2019.csv')
```

```python
# Draw a heatmap in order to have a first overview of
#the correlation between severals parameters

# create a copy our data frame to work on it
df_GEO_HM = df_GEO.copy()
# delete the columns 'Country' in order to use our data frame
del df_GEO_HM['Country']

# creat a correlation matrix 
corr = df_GEO_HM.corr()
corr = corr.abs()
```

```python
# creat a mask to hide the top the heatmap
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True

# creat the heatmap
sns.heatmap(corr, mask=mask, annot=True, linewidths=.5, cmap='Reds')
```
