```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt


# reads our data set and creats a pandas dataframe
DF = pd.read_csv('DATA/Democracy-Index.csv', usecols=range(1,10))
# columns explicitation : 
    # "Country" : name of the country
    # "Democracy": 1 is a democracy, 0 a dictatorship (The Economist gave an Overall score.
    # Here, if this score is between 6 and 10 we consider the country a democracy, if it is between 0 and 6, a dictature.)
    # "Expectancy": age expectancy in year
    # "PopulationD" : density of population in hab/km2
    # "Gini" : coefficient de Gini, evaluates wealth inequality
    # "AgeMed" : median age of the population in year
    # "Skyscraper" : high the highest building in meters 
    # "Children": number of children per woman
    # "PressF" : press freedom score: 0 = great freedom, 100 = no freedom
```

```python
# Draw a heatmap in order to have a first overview of
#the correlation between severals parameters

# delete the columns 'Country' in order to use our data frame
del DF['Country']

# creat a correlation matrix 
corr = DF.corr()
corr = corr.abs()
```

```python
# creat a mask to hide the top the heatmap
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True

# creat the heatmap
sns.heatmap(corr, mask=mask, annot=True, linewidths=.5, cmap='Reds')
```

```python
# create two data frames to split by democracy or dictatorship
# and draw for each parameter the density of distribution

dictature = DF.loc[DF['Democracy']==0]
democracy = DF.loc[DF['Democracy']==1]
```
